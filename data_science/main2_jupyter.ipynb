```python
import pandas as pd  # type: ignore
import matplotlib.pyplot as plt # type: ignore | # É uma biblioteca utilizada para visualizar dados, como gráficos, diagramas e figuras personalizadas 
import numpy as np  # type: ignore
```

```python
gas = pd.read_csv('gas/gas_prices.csv')
gas
```

```python
# CRIANDO O PRIMEIRO GRÁFICO USANDO O plt.plot


plt.title('Price in US$') # Com o plt.title podemos adicionar título

plt.plot(gas['Year'], gas['Australia'], 'g.--',label = 'Australia') # Com o label é possível adicionar uma legenda
plt.plot(gas['Year'], gas['Italy'], 'b.-',label = 'Italy')
plt.plot(gas['Year'], gas['USA'], 'r.--',label = 'USA') # 'r.--'  (r = Cor | . = Estilo do Marcador| -- = Estilo da Linha)
plt.plot(gas['Year'], gas['UK'], 'y.-',label = 'UK')

plt.xticks(gas['Year'][::2]) # Usado para personalizar os Valores e Rótulos que serão exibidos no eixo X

plt.xlabel('Years') # Adiciona um rótulo no eixo X
plt.ylabel('US$') # Adiciona um rótulo no eixo Y

plt.legend() # Usado para aparecer a Legenda

plt.savefig('country_semforloop.pdf') # Usado para salvar o Gráfico | OBS: Deve ser colocado antes do " plt.show() "

plt.show()
```

```python
plt.title('Price in US$') # Com o plt.title podemos adicionar título

# CRIANDO GRÁFICOS USANDO O (For Loop)

# Lista: 
list_country = ['Australia', 'USA', 'Italy', 'UK', 'Mexico']

for country in gas:
    if country in list_country:
        # Características:
        plt.plot(gas['Year'], gas[country], label = [country], marker='.') # marker = Altera o estilo dos Marcadores

plt.xticks(gas['Year'][::2]) # Usado para personalizar os Valores e Rótulos que serão exibidos no eixo X

plt.xlabel('Years') # Adiciona um rótulo no eixo X
plt.ylabel('US$') # Adiciona um rótulo no eixo Y

plt.legend() # Usado para aparecer a Legenda

plt.savefig('country_comforloop.png') # Usado para salvar o Gráfico | OBS: Deve ser colocado antes do " plt.show() "

plt.show()

```
